```python
import os
import glob
import rasterio
import numpy as np
import torch
import random
import xarray as xr
```

```python
def gerador(pasta_pre_processada, batch_size=8, embaralhar=True):

    arquivos_pt = glob.glob(os.path.join(pasta_pre_processada, '*.pt'))
    
    if embaralhar:
        random.shuffle(arquivos_pt)
        
    lote_imagens = []
    lote_mascaras = []
    
    for arquivo_path in arquivos_pt:
        dados_salvos = torch.load(arquivo_path)
        
        lote_imagens.append(dados_salvos['imagem'])
        lote_mascaras.append(dados_salvos['mascara'])
        
        if len(lote_imagens) == batch_size:
            yield torch.stack(lote_imagens), torch.stack(lote_mascaras)
            lote_imagens = []
            lote_mascaras = []
            
    if len(lote_imagens) > 0:
        yield torch.stack(lote_imagens), torch.stack(lote_mascaras)
```

```python

pasta_pre_processada = r"dataset/pre_processado"

gerador = gerador(pasta_pre_processada, batch_size=8)

print("Testando o gerador rápido...")
for batch_img, batch_mask in gerador:
    print("Batch Imagens:", batch_img.shape)
    print("Batch Máscaras:", batch_mask.shape)
```
